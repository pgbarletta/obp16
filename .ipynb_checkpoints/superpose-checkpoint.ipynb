# Difference vectors projections onto PCA subspace 
Este código toma una trayectoria completa (previamente fitteada al average) y obtiene los vectores diferencia normalizados. Luego lee los PCA (q también fueron obtenidos en base al average) y proyecta todas los vectores diferencia en la base de los PCA.

```julia
using DataFrames
using Chemfiles
using Gadfly
using Cairo
using Distributions
```

```julia
function read_ptraj_modes(file, modes_elements)    
    modes_file=open(file, "r")
    modes_text = readdlm(modes_file, skipstart=0, skipblanks=true, 
    ignore_invalid_chars=true, comments=true, comment_char='\*')
    close(modes_file)

    nmodes = modes_text[1, 5]
    ncoords = convert(Int64, modes_elements)
    lines = ceil(Int64, ncoords/7)
    rest = convert(Int64, ncoords % 7)
    
    eval=Array{Float64}(nmodes);
    mode = Array{Float64}(ncoords, nmodes);
    temp1=Array{Float64}(ncoords, 1);

    j=lines + 1 + 2 # 1 p/ q lea la prox linea 2 por el header

    for i=1:nmodes
        eval[i] = modes_text[j, 2]
        temp = transpose(modes_text[(j+1):(lines+j), :])
        mode[:, i] = reshape(temp, ncoords)
        j = j + lines + 1
    end
    
    for i=1:nmodes
        mode[: ,i] = mode[:, i] / norm(mode[:, i])
    end
    
    return mode, eval
end
```

## wt_apo

```julia
# Preparo variables

wt_apo_tr=Trajectory("/home/german/labo/16/obp/run/wt_apo/avg_fit_wt_apo.nc")
wt_apo_atom_cnt=convert(Int64, natoms(read_step(wt_apo_tr, 1)))
wt_apo_frames_cnt=convert(Int64, nsteps(wt_apo_tr))

diff_vtr_wt_apo = Array{Float32}(357, wt_apo_frames_cnt-1);
diff_vtr_wt_apo_out = Array{Float32}(wt_apo_atom_cnt*3);
modes_wt_apo = Array{Float32}(357, 351);
temp=Array{Int16}(119, 1);

# Acá leo los indices de los carbonos alfa p/ subsettear los vectores posición.
file_wt_apo_ca_index=open("/home/german/labo/16/obp/run/wt_apo/ca_index_wt_apo");
temp = readdlm(file_wt_apo_ca_index);
close(file_wt_apo_ca_index);
temp = convert(Array{Int16,2}, temp);

wt_apo_ca_index = Array{Int16, 1}(357)
wt_apo_ca_index1 = Array{Int16, 1}(119)
wt_apo_ca_index1 = temp[:, 1]
wt = Array{Int16, 1}(357)
for i in 1:119
    j=i*3
    wt_apo_ca_index[j-2] = wt_apo_ca_index1[i] * 3 - 2
    wt_apo_ca_index[j-1] = wt_apo_ca_index1[i] * 3 - 1
    wt_apo_ca_index[j] = wt_apo_ca_index1[i] * 3
end

# Obtengo los vectores diferencia de los Calpha entre c/ snapshot
for i=1:wt_apo_frames_cnt-1
    wt_apo_fr_prev = read_step(wt_apo_tr, 0)
    wt_apo_prev_xyz = getindex(positions(wt_apo_fr_prev), wt_apo_ca_index)
    wt_apo_fr = read_step(wt_apo_tr, i)
    wt_apo_xyz = getindex(positions(wt_apo_fr), wt_apo_ca_index)
    diff_vtr_wt_apo[:, i]= wt_apo_xyz - wt_apo_prev_xyz
    diff_vtr_wt_apo[:, i]= diff_vtr_wt_apo[:, i] / norm(diff_vtr_wt_apo[:, i])
end

# Leo los modos de PCA
modes_wt_apo, evals_wt_apo = read_ptraj_modes("run/wt_apo/modes_wt_apo", 357);

# Proyecto los vectores diferencia sobre la base de PCA y obtengo los p-numbers
projection = transpose(modes_wt_apo) * diff_vtr_wt_apo;
projection1 = projection.^4;
pnum_wt_apo = mapslices(sum, projection1, 1);
pnum_wt_apo = map(x -> 1/x, pnum_wt_apo);
```
